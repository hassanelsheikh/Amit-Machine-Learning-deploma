```python
x=df[['ENGINESIZE']]
y=df['CO2EMISSIONS']
```

```python
from sklearn import linear_model
```

```python
model = linear_model.LinearRegression()
model.fit(x,y)
model.predict([[5]])[0] 
```

```python
#Clculate mean squared error
from sklearn.metrics import mean_squared_error
```

```python
from numpy.ma.core import sqrt
y_pred = model.predict(x)
mse=mean_squared_error(y,y_pred)
print(sqrt(mse))#square root of mean squared error
print(model.intercept_) #a0
print(model.coef_)#a1

```

```python
plt.figure(figsize=(10,8))
plt.scatter(x,y)
plt.plot(x,y_pred, c='r')
```

```python
#LETS SPLIT TEST AND TRAIN
from sklearn.model_selection import train_test_split
X = df[['FUELCONSUMPTION_COMB']].values
Y = df['CO2EMISSIONS']

X_train, X_test, y_train, y_test = train_test_split(X,Y,test_size=0.33, random_state=30)
model_x = linear_model.LinearRegression()
model_x.fit(X_train,y_train)
y_new = model_x.predict(X_test)

plt.scatter(X_train,y_train)

plt.scatter(X_test,y_test)

plt.plot(X_test, y_new, c='r') #Linear regression model using test and trainW
```

```python
#try split train in the previous linear regression models
x=df[['ENGINESIZE']].values
y=df['CO2EMISSIONS'].values
x_train, x_test, y_train, y_test = train_test_split(x,y,train_size=0.33,random_state=2)
```

```python
plt.figure(figsize=(10,8))
model = linear_model.LinearRegression()
model.fit(x_train,y_train)
y_n = model.predict(x_test)

plt.scatter(x_train, y_train)

plt.plot(x_test,y_n, c='r')
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test, y_n) #888 mean squared error

```

```python
#!pip install mlxtend --upgrade

from mlxtend.evaluate import bias_variance_decomp

avg_loss, avg_bias, avg_var = bias_variance_decomp(model, x_train, y_train,x_test,y_test,loss = 'mse', random_seed = 32, num_rounds = 200)
print(avg_loss)
print(avg_bias)
print(avg_var)
```
